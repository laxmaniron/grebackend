```python

# import openpyxl module 
import openpyxl 
  
# Call a Workbook() function of openpyxl  
# to create a new blank Workbook object 
wb = openpyxl.Workbook() 
  
# Get workbook active sheet   
# from the active attribute 
sheet = wb.active 
  
all_objects=source.find()

cell_row=3

for word in all_objects:
    
    # simple_word
    type_list=[]
    meaning_list=[]
    word_example=[]
    
    
    print(word["wordname"])
    print(word["typeofvocab"])
    print(word["word_meaning"])
    print(word["word_mnemonic"])
    print(word["word_example"])
    print(word["origin_of_word"])
    
    print("/n/n")
    
    word_column = sheet.cell(row = cell_row, column = 1)
    word_column.value = word["wordname"]
    
    type_column = sheet.cell(row = cell_row, column = 2)
    type_list.append(word["typeofvocab"])
    type_column.value = str(type_list)
    
    meaning_column = sheet.cell(row = cell_row, column = 3)
    meaning_list.append(word["word_meaning"])
    meaning_column.value = str(meaning_list)
    
    mnemonic_column = sheet.cell(row = cell_row, column = 4)
    mnemonic_column.value = word["word_mnemonic"]
    
    example_column = sheet.cell(row = cell_row, column = 5)
    word_example.append(word["word_example"])
    example_column.value = str(word_example)
    
    origin_column = sheet.cell(row = cell_row, column = 6)
    origin_column.value = word["origin_of_word"]
    
    cell_row += 1
  

wb.save("./demo.xlsx") 
```

```python
import openpyxl
from openpyxl import Workbook, load_workbook
import pymongo

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["GREnew"]
source = mydb["grenewvocabs"]

# path = "./apparels_maininfo.xlsx"
path = "./demo.xlsx"

wb_obj = openpyxl.load_workbook(path)

sheet_obj = wb_obj.active

inserter = {}

p = 0

count = 0


for dress in range(3, (sheet_obj.max_row)+1):
    
    inserter = {}
    
    m = sheet_obj.cell(row=dress, column=1)

    inserter["wordname"] = m.value
    
    type_v = sheet_obj.cell(row=dress, column=2)

    inserter["typeofvocab"] = eval(type_v.value)

    meaning = sheet_obj.cell(row=dress, column=3)
    inserter["word_meaning"] = eval(meaning.value)

    mnemonic = sheet_obj.cell(row=dress, column=4)
    inserter["word_mnemonic"] = mnemonic.value

    word_example = sheet_obj.cell(row=dress, column=5)
    inserter["word_example"] = eval(word_example.value)
    
    
    origin = sheet_obj.cell(row=dress, column=6)
    inserter["origin_of_word"] = origin.value
    source.insert_one(inserter)

    print(inserter)
```
